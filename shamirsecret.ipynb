```python
k = 2
```

```python
n = 6
```

```python
secret = 1234
```

```python
a = random.randint(-5, 5)
print("Nombre aléatoire a =", a)
```

```python
parts = []
for x in range(1, n + 1):
    y = secret + a * x 
    parts.append((x, y)) 
```

```python
print("\nParts distribuées :")
for part in parts:
    print(part)
```

```python
x1, y1 = parts[1]
x2, y2 = parts[4]
```

```python
a = (y2 - y1) / (x2 - x1)
print(" retrouvée a =", a)
```

```python
s = y1 - a * x1
print("Secret retrouvé :", s)
```
